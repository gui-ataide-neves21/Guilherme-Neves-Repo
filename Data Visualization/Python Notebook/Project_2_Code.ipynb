```python
import pandas as pd
import openpyxl
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

```python
covid_data = pd.read_excel('owid-covid-data.xlsx', engine='openpyxl')

# Select only the European countries
covid_data_europe = covid_data[covid_data['continent'] == 'Europe']
```

```python
covid_data_europe['date'] = pd.to_datetime(covid_data_europe['date'], format='%Y-%m-%d')
```

```python
covid_data_europe.drop('tests_units',axis=1, inplace=True)
```

```python
covid_data_europe[covid_data_europe.iloc[:,4:] < 0] = np.nan 
```

```python
mean_europe = covid_data_europe.groupby('date').sum().reset_index()
```

```python
covid_data_europe
```

```python
covid_data_europe = pd.concat([covid_data_europe, mean_europe], axis=0)
```

```python
covid_data_europe.to_excel('covid_data_europe.xlsx')
```
